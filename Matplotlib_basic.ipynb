1) Plotting :-
* The plot() function is used to draw points (markers) in a diagram.
* By default, the plot() function draws a line from point to point.
* The function takes parameters for specifying points in the diagram.

```python
# We can also use dot instead of line
x = [4,6,2,6,4,5]
y = [10,15,3,6,13,3]
plt.plot(x,y,'o')
plt.show()
```

```python
# default it will either start from 0 and 1
y = [10,15,3,6,13,3]
plt.plot(y)
plt.show()
```

mfc =  marker face color,
mec = marker h color,
ls = linestyle,
lw = linewidth,
ms = markersize

```python
x = [4,6,2,6,4,5]
y = [10,15,3,6,13,3]
plt.plot(x,y, marker='*', ls='-.', lw=2, ms = 19, mfc = 'DarkMagenta', mec = 'Gold', color = 'DeepPink')
plt.xlabel('salary', c = 'b')
plt.ylabel('Bike',c = 'b')
plt.title('Company')
plt.show()
```

```python
x = [4,6,2,6,4,5]
y = [10,15,3,6,13,3]
plt.plot(x,y, marker='*', ls ='-.', ms = 19, mfc = 'DarkMagenta', mec = 'Gold', color = 'DeepPink')
plt.xlabel('salary', c = 'b')
plt.ylabel('Bike',c = 'b')
plt.title('Company')
plt.grid(color = 'r',linestyle = '--')
plt.show()
```

2) Scatter diagram :-
* The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
age = [3,5,6,8,2,9,11,2,3,7]
speed = [98,87,75,67,121,45,40,102,87,68]
speed1 = [89,69,83,75,110,64,50,99,81,78]
```

```python
plt.scatter(age,speed, marker = 's', color ='b')
plt.scatter(age,speed1, marker = 'h', color ='purple')
plt.show()
```

```python
plt.scatter(age,speed, marker = 's', color ='b')
plt.scatter(age,speed1, marker = 'h', color ='purple')
plt.xlabel("Age", color ='m',)
plt.ylabel("Speed", color = 'm')
plt.title("Car Comparision", color = 'b')
plt.legend(['speed','speed1'],ncol = 2, facecolor = "c", fontsize = 15 )
plt.show()
```

3) Bar diagram :-
* The bar() function takes represented by the first and second argument as arrays.

```python
x = ['A','B','C','D']
y = [6,9,3,10]
plt.bar(x,y, width = 0.5, align ='edge')
plt.show()
```

* If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
x = ['A','B','C','D']
y = [6,9,3,10]
plt.barh(x,y, height= 0.5, color ="m")
plt.show()
```

5) Pie chart :-
* As you can see the pie chart draws one piece for each values in the array
* By default the plotting of the first wedge start from the x-axis and moves counterclockwise.

```python
y = [30,20,10,40]
x = ['A','B','C','D']
c = ['c','m','DeepPink','b' ]
plt.pie(y, labels = x, colors = c)
plt.show()
```

```python
# The explode parameter, if specified and not none, must be an array with one values for each wedge (0.0,0.1,..)
y = [30,20,10,40]
x = ['A','B','C','D']
c = ['c','m','DeepPink','b' ]
plt.pie(y, labels = x, colors = c, explode = [0.0,0.4,0.0,0.0], shadow = True)
plt.show()
```
