* fillna
* To fill the blank values

```python
# To fill the blank values
df.fillna(123)
df.head()
```

```python
# To fill perticular one column
df['Date'].fillna(2022)
```

* mean

```python
# To replace null values from mean
cal_mean = df['Calories'].mean()
cal_mean
```

```python
df['Calories'].fillna(cal_mean)
```

* median

```python
# To replace null values from median
cal_median = df['Calories'].median()
cal_median
```

* mode

```python
# To replace null values from mode
cal_mode = df['Calories'].mode()
cal_mode
```

```python
# To formatting the column
df['date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
# To replace the values of perticular column
df['Duration'].replace(60,30)
```

```python
df.reset_index(inplace = True)
df.head()
```

```python
df.drop(['Date'],axis = 1, inplace = True)
df.head()
```
